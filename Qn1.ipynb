```python
import pandas as pd

# Replace with your exact uploaded filename if different
filename = 'API_SP.POP.TOTL_DS2_en_csv_v2_127006.csv'

# Skip metadata rows at the top
df = pd.read_csv(filename, skiprows=4)

# Show first few rows
df.head()
```

```python
import matplotlib.pyplot as plt

# Choose a recent year, e.g., 2022
year = '2022'

# Filter and clean data
data = df[['Country Name', year]].dropna()
data[year] = data[year].astype(int)

# Get top 10 countries by population
top10 = data.sort_values(by=year, ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
plt.bar(top10['Country Name'], top10[year], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title(f'Top 10 Most Populous Countries in {year}')
plt.ylabel('Population')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Log-transform to handle wide range
data['log_pop'] = np.log10(data[year])

# Plot histogram
plt.figure(figsize=(10,6))
plt.hist(data['log_pop'], bins=30, edgecolor='black', color='orange')
plt.xlabel('Log10(Population)')
plt.ylabel('Number of Countries')
plt.title(f'Population Distribution in {year} (log scale)')
plt.grid(True)
plt.show()
```
